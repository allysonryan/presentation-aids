```python
from skimage import measure
from skimage import filters
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
n = 12
l = 256
np.random.seed(1)
im = np.zeros((l, l))
points = l * np.random.random((2, n ** 2))
im[(points[0]).astype(int), (points[1]).astype(int)] = 1
im = filters.gaussian(im, sigma= l / (4. * n))
blobs = im > 0.7 * im.mean()
labels = measure.label(blobs, background=0, connectivity=2)

plt.figure(figsize=(3,3))
plt.imshow(blobs, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.savefig('binary_blobs.svg', dpi = 300)
plt.show()

plt.figure(figsize=(3,3))
plt.imshow(labels, cmap='gnuplot')
plt.axis('off')
plt.tight_layout()
plt.savefig('labeled_blobs.svg', dpi = 300)
plt.show()
```
